```python
from muestreador import MuestreoINE
m = MuestreoINE(tamaño_poblacion=427812)
import pandas as pd
import numpy as np
```

```python
error_y_pob = []
def calculo_error(varianza, media):
    if  varianza < 0.06:
        error = 0.03
    elif 0.06 <= varianza < 0.2:
        error = 0.05
    elif 0.2 <= varianza:
        error = 0.1
    error_y_pob.append((error, min(250, max(m.tamaño_srs(confianza= 0.95, varianza= varianza, error= error * media), 40))))
```

```python
df = pd.read_excel("Medias y varianzas.xlsx")
df.head(3)
```

```python
poblaciones = []
var_y_medias = zip(df["VARIANZA"].tolist(), df["MEDIA"].tolist())
for var, med in var_y_medias:
    calculo_error(var, med)

error_y_pob
```
